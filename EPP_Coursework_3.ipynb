```python
!pip install astropy
from astropy.io import ascii
import numpy as np
from matplotlib import pyplot as plt
```

```python
dataarray = ascii.read("https://raw.githubusercontent.com/Hamza-Suleman/EPP-Coursework3/main/datafile.csv")
print(dataarray)
```

###Calculating m12

```python
mwlist = []

for line in dataarray:
    # 3 possibilities to pair  2 light jets:

    m12 = np.sqrt((line[13]+line[17])**2 - (line[14]+line[18])**2 - (line[15]+line[19])**2 - (line[16]+line[20])**2)/1000

    if abs(m12-80)<40:
            mwlist.append(m12)

m12mean = np.mean(mwlist)
print('Mean: ', m12mean)
          
plt.hist(mwlist, bins=20, range=(40,120), label="m(W) [GeV]") 
plt.xlabel("m(W) [GeV]")
plt.savefig('MW.png')
plt.show()
```
