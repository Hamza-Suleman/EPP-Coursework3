```python
!pip install astropy
from astropy.io import ascii
import numpy as np
from matplotlib import pyplot as plt
import statistics
```

```python
dataarray = ascii.read("https://raw.githubusercontent.com/Hamza-Suleman/EPP-Coursework3/main/datafile.csv")
print(dataarray)
```

```python
test = []
for i in range(len(dataarray)):
  a = np.sqrt((dataarray[i]["J1_E"]+dataarray[i]["J2_E"])**2 - (dataarray[i]['J1_Px']+dataarray[i]['J2_Px'])**2 - (dataarray[i]['J1_Py']+dataarray[i]['J2_Py'])**2 - (dataarray[i]['J1_Pz']+dataarray[i]['J2_Pz'])**2)/1000
  if abs(a-80)<40:
    test.append(a)
print(np.mean(test))
```

###Calculating m12

```python
mwlist = []

for line in dataarray:
    # 3 possibilities to pair  2 light jets:

    m12 = np.sqrt((line[13]+line[17])**2 - (line[14]+line[18])**2 - (line[15]+line[19])**2 - (line[16]+line[20])**2)/1000

    if abs(m12-80)<40:
            mwlist.append(m12)

m12mean = np.mean(mwlist)
print('Mean: ', m12mean)
print("Standard Deviation: ", (statistics.stdev(mwlist)/100)*m12mean)
          
plt.hist(mwlist, bins=20, range=(40,120), label="m(W) [GeV]") 
plt.xlabel("m(W) [GeV]")
plt.savefig('MW.png')
plt.show()
```

```python
firstcomb = []

for i in range(len(dataarray)):
  B1B2 = np.sqrt(abs((dataarray[i]["B1_E"]+dataarray[i]["B2_E"])**2 - (dataarray[i]['B1_Px']+dataarray[i]['B2_Px'])**2 - (dataarray[i]['B1_Py']+dataarray[i]['B2_Py'])**2 - (dataarray[i]['B1_Pz']+dataarray[i]['B2_Pz'])**2))/1000
  J1J2 = np.sqrt(abs((dataarray[i]["J1_E"]+dataarray[i]["J2_E"])**2 - (dataarray[i]['J1_Px']+dataarray[i]['J2_Px'])**2 - (dataarray[i]['J1_Py']+dataarray[i]['J2_Py'])**2 - (dataarray[i]['J1_Pz']+dataarray[i]['J2_Pz'])**2))/1000
  Muon = np.sqrt(abs((dataarray[i]["mu_E"])**2 - (dataarray[i]['mu_Px'])**2 - (dataarray[i]['mu_Py'])**2 - (dataarray[i]['mu_Pz'])**2))/1000
  sum = B1B2 + J1J2 + Muon
  if 100<sum<250:
    firstcomb.append(sum)

firstcombination = np.mean(firstcomb)
#print(firstcomb)
print('Mean: ', firstcombination)
print("Standard Deviation: ", np.std(firstcomb))
          
plt.hist(firstcomb, bins=25, range=(100,250), label="m(T) [GeV]") 
plt.title("A range of top quark values for a combination of B1B2, J1J2 and a Muon")
plt.ylabel("Arbitrary Units")
plt.xlabel("m(T) [GeV]")
plt.show()
```
