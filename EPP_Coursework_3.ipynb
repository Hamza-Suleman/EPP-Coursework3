```python
!pip install astropy
from astropy.io import ascii
import numpy as np
from matplotlib import pyplot as plt
import statistics
```

```python
dataarray = ascii.read("https://raw.githubusercontent.com/Hamza-Suleman/EPP-Coursework3/main/datafile.csv")
print(dataarray)
```

```python
test = []
for i in range(len(dataarray)):
  a = np.sqrt((dataarray[i]["J1_E"]+dataarray[i]["J2_E"])**2 - (dataarray[i]['J1_Px']+dataarray[i]['J2_Px'])**2 - (dataarray[i]['J1_Py']+dataarray[i]['J2_Py'])**2 - (dataarray[i]['J1_Pz']+dataarray[i]['J2_Pz'])**2)/1000
  if abs(a-80)<40:
    test.append(a)
print(np.mean(test))
```

###Calculating m12

```python
mwlist = []

for line in dataarray:
    # 3 possibilities to pair  2 light jets:

    m12 = np.sqrt((line[13]+line[17])**2 - (line[14]+line[18])**2 - (line[15]+line[19])**2 - (line[16]+line[20])**2)/1000

    if abs(m12-80)<40:
            mwlist.append(m12)

m12mean = np.mean(mwlist)
print('Mean: ', m12mean)
print("Standard Deviation: ", (statistics.stdev(mwlist)/100)*m12mean)
          
plt.hist(mwlist, bins=20, range=(40,120), label="m(W) [GeV]") 
plt.xlabel("m(W) [GeV]")
plt.savefig('MW.png')
plt.show()
```

###Muon Mass

```python
muon_mass_list = []
for i in range(len(dataarray)):
  a = np.sqrt(abs(((dataarray[i]["mu_E"])**2 - (dataarray[i]['mu_Px'])**2 - (dataarray[i]['mu_Py'])**2 - (dataarray[i]['mu_Pz'])**2)/1000))
  muon_mass_list.append(a)
muon_mass = np.mean(muon_mass_list)
print("Muon Mass: ", muon_mass)
print("Standard Deviation: ", (statistics.stdev(muon_mass_list)/100)*muon_mass)
```
